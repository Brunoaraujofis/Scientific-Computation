# Container Types 

Este capítulo trata de Container Types. 
Container Types são usado para agrupar objetos. A principal diferença entre Container Types esta  na forma que os elementos individuais podem ser acessados e nas operações que podem ser definidas. 


# Listas

Uma lista é propriamente como o nome diz, definimos a seguir duas listas;

```python
L = ['a', 20.0, 5]
M = [3, ['a', -3.0, 5]]
```

Os elementos individuais são idexados por 0,1,...
O comando L[i] , i = 0,1,2,... chama o i-ésimmo elemento da lista e.g. L[0] = 'a', 
M[1] = ['a',-3.0, 5]

```python
L[0], M[1], L[2]
```

A lista M possui uma lista dentro dela. Com dois indices podemos acessar os elementos dentro da lista interior à M, e.g.,

```python
M[1][0]
```

Uma lista com subsequentes números inteiros pode ser gerada facilmente pelo comando list(range):

```python
L1 = list(range(10))
L1
```

Outra forma de gerar lista pode ser dando o começo, fim e o passo da lista, e.g.,

```python
L2 = list(range(10,30,2))
L2
```

Podemos ver o tamanho (número de elementos) de uma lista pelo comando len:

```python
len(L2), len(L1)
```

# SLICING 

Slicing ou repartir uma lista entre os elementos i e j cria uma nova lista onde o primeiro elemento da nova lista é o i-ésimo elemento da lista antiga e o último elemento é o (j-1)-ésimo  elemento da lista antiga. 
O comando L[i:j] cria uma nova lista pegando do i-ésimo elemento até o (j-1)-ésimo elemento

```python
L1[2:7], L1[:7],L1[2:-1], L1[:-2]
```

l[i:] significa remova os i-ésimos primeiros elementos, l[:i] significa pegue
os i-ésimos elementos, já l[:-i] significa remova os i-ésimos elementos finais
e l[-i:] pegue os i-ésimos elementos finais. Podemos combinar os comando por 
l[i:] and l[:-j] significando remova os i-ésimos elementos iniciais e também 
remova os j-ésimos elementos finais. 

```python
Lc = ['C','l','o','u','d','s']
Lc[1:5],Lc[1:], Lc[:]
```

```python
a = [1,2,3]
for i in range(4):
        print(np.sum(a[0: i-1]))
```

# Strides 

Strides ou passos servem para especificar os passos quando estamos computando slices (fatias). Segue alguns exemplos de Strides usado em Slices.

```python
L = list(range(100))
L[:10:2], L[::20], L[10:20:3]
```

Note que podemos usar Strides negativos. Lembre que definimos L1 por list(range(10)). Assim, podemos inverter a ordem de uma lista pelo comando a seguir:

```python
L[20:10:-3], L1[::-1]
```
